```python
def tell_distr(col):
    min_data = col.min()
    max_data = col.max()
    mean_data = col.mean()
    mode_data = col.mode()
    median_data = col.median()
    print(f"The distributions are: \n minimum: {min_data} \n maximum: {max_data} \n mean: {mean_data} \n mode: {mode_data} \n median: {median_data}")
    
```

```python
tell_distr(flights_df.DepDelay)
```

```python
tell_distr(flights_df.ArrDelay)
```
